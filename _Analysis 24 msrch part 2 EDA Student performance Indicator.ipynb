## Explore More Visualization

```python
fig,axis=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=da,x='average_score',bins=30,kde=True,color='green')

plt.subplot(122)
sns.histplot(data=da,x='average_score',bins=30,kde=True,hue='gender')
```

## insights

Female student tend to perform well than male students

```python
plt.subplots(1,3,figsize=(25,10))
plt.subplot(141)
sns.histplot(data=da, x='average_score', kde=True, hue="lunch")
plt.subplot(142)
sns.histplot(data=da[da.gender=='female'], x='average_score', kde=True, hue ="lunch")


plt.subplot(143)
sns.histplot(data=da[da.gender=='male'], x='average_score', kde=True, hue ="lunch")

```

## insights 
1) Standard Lunch  help student perform well in exams
2) Standard Lunch helps perform well in exams be it a male of female 

```python
plt.subplots(1,4,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=da,x='average_score',kde=True,hue='parental level of education')

plt.subplot(142)
sns.histplot(data=da[da.gender=='male'],x='average_score',kde=True,hue='parental level of education')


plt.subplot(143)
sns.histplot(data=da[da.gender=='female'],x='average_score',kde=True,hue='parental level of education')





```

## Insights 

1) In gender parent's education don't help student perform well in exam 
2) 2nd plot show that parent's whose education is of associate's ddegree or master's degree their  male child tend to perform well in exam  
3) 3rd plot we can see there is no effect of parent's  education  on female  students
