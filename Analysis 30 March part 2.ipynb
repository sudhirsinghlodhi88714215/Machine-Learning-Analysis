```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Cleaning

## Missing Values

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,'Regsion']=0
dataset.loc[122:,"Regsion"]=1
df=dataset
```

```python
df
```

```python
df.info()
```

```python
df.head()
```

```python
df[["Regsion"]]=df[["Regsion"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

## Removing the Null Values

```python
df=df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

## Changing the other columns to float data datatype

```python
df.columns
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset_UPDATE.csv',index=False)
```
